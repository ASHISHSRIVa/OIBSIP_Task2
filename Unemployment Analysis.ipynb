# Estimated Unemployment Rate

```python
dfr = round(dfr.groupby(['State']).mean().reset_index(),6)
dfr
```

```python
st = dfr[["State","Estimated Unemployment Rate (%)"]].groupby("State").sum().sort_values(by="Estimated Unemployment Rate (%)", ascending  =False)
st.head(28)
```

```python
sta = dfr[["State","Estimated Employed"]].groupby("State").sum().sort_values(by="Estimated Employed", ascending =False)
sta.head(28)
```

```python
fig=plt.figure()
ax1=fig.add_subplot(1,3,3)
ax5=fig.add_subplot(1,3,1)


#Unemployed
st[:10].plot(kind="bar",color="yellow",figsize=(30,10),ax=ax1)
ax0.set_title("People Unemployed in each State")
ax0.set_xlabel("State")
ax0.set_ylabel("Number of People Unemployed (in %)")

#Employed
sta[:10].plot(kind="bar",color="yellow",figsize=(30,10),ax=ax5)
ax0.set_title("People Employed in each State")
ax0.set_xlabel("State")
ax0.set_ylabel("Number of People Employed (in %)")
```
